## Задание 4

```python
text = open("text.txt", "r")
d = dict()

for line in text:
    line = line.strip()
    line = line.lower()
    words = line.split(" ")
    for word in words:
        if word in d:
            d[word] = d[word] + 1
        else:
            d[word] = 1

for key in list(d.keys()):
    print(key, ":", len(key))
```

```python
import random
import numpy as np
A = np.random.randint(0,9, size=(8,8))
b = [sum(x) for x in zip(*A)]
print(A)
k = b.index(min(b))
print('-------------------')
for i in range(8):
    min_col = A[i][k]
    print(min_col)
print('-------------------')
print('Медиана numpy:', np.median(min_col))

def my_median(lst):
    n = 8
    s = np.sort(lst)
    return (sum(s[n//2-1:n//2+1])/2.0, s[n//2])[n % 2] if n else None

print('Медиана расчётным способом:', my_median(min_col))
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('films.csv')

print('Строк и столбцов:',df.shape) # форма датасета
print('-------------------')
print('Типы данных по столбцам:\n', df.dtypes) # типы данных
print('-------------------')

# отбор числовых колонок
df_numeric = df.select_dtypes(include=[np.number])
numeric_cols = df_numeric.columns.values
print('Числовые колонки:', numeric_cols)
print('-------------------')

# отбор нечисловых колонок
df_non_numeric = df.select_dtypes(exclude=[np.number])
non_numeric_cols = df_non_numeric.columns.values
print('Нечисловые колонки:',non_numeric_cols)
print('-------------------')
new_df = df[['imdb_id', 'director', 'budget', 'revenue']].dropna()
new_df['director'] = new_df['director'].str.lower()
director_df = pd.DataFrame(new_df['director'].value_counts(dropna=False))
director_df['more'] = new_df[new_df['revenue'] > new_df['budget']]['director'].value_counts(dropna=False)
director_df = director_df.dropna()
director_df['percent'] = director_df['more'] / director_df['director']
print('Режиссеры, у которых все фильмы собрали больше, чем бюджет: \n', '\n'.join(director_df[director_df.percent == 1].index))
```
